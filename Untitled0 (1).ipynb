## 2. Missing Data Analysis

### Identifying Missing Values

```python
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0]
missing_percentage = (missing_data / len(df)) * 100

missing_info = pd.DataFrame({
    'Missing Count': missing_data,
    'Missing Percentage (%)': missing_percentage
})
display(missing_info.sort_values(by='Missing Percentage (%)', ascending=False))
```

### Visualizing Missing Values

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

### Missing Data Summary

From the analysis, the columns with missing values are:
- `Income`: 39 missing values
- `Credit_Score`: 2 missing values
- `Loan_Balance`: 29 missing values

The percentages are relatively low, suggesting that imputation might be a viable strategy rather than dropping these columns. The heatmap visually confirms the presence and distribution of these missing values.

## 3. Key Findings and Risk Indicators

### Target Variable Distribution

```python
delinquency_counts = df['Delinquent_Account'].value_counts()
delinquency_percentage = df['Delinquent_Account'].value_counts(normalize=True) * 100

delinquency_info = pd.DataFrame({
    'Count': delinquency_counts,
    'Percentage (%)': delinquency_percentage
})
delinquency_info.index = ['Non-Delinquent (0)', 'Delinquent (1)']
display(delinquency_info)

plt.figure(figsize=(6, 4))
sns.barplot(x=delinquency_info.index, y='Percentage (%)', data=delinquency_info, palette='coolwarm')
plt.title('Distribution of Delinquent Accounts')
plt.ylabel('Percentage')
plt.show()
```

### Correlation with Delinquency (Numeric Features)

```python
correlation_with_delinquency = df.corr(numeric_only=True)['Delinquent_Account'].sort_values(ascending=False)
display(correlation_with_delinquency)

plt.figure(figsize=(8, 6))
sns.barplot(x=correlation_with_delinquency.index, y=correlation_with_delinquency.values, palette='viridis')
plt.title('Correlation of Numeric Features with Delinquent_Account')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Correlation Coefficient')
plt.tight_layout()
plt.show()
```

### Categorical Feature Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

sns.countplot(x='Employment_Status', hue='Delinquent_Account', data=df, ax=axes[0], palette='pastel')
axes[0].set_title('Delinquency by Employment Status')
axes[0].tick_params(axis='x', rotation=45)

sns.countplot(x='Credit_Card_Type', hue='Delinquent_Account', data=df, ax=axes[1], palette='pastel')
axes[1].set_title('Delinquency by Credit Card Type')
axes[1].tick_params(axis='x', rotation=45)

sns.countplot(x='Location', hue='Delinquent_Account', data=df, ax=axes[2], palette='pastel')
axes[2].set_title('Delinquency by Location')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### Key Findings Summary

- **Target Variable Imbalance**: **16% of customers are delinquent**, indicating a class imbalance that will need to be addressed during model training.
- **Numeric Feature Correlation**:
    - `Income`, `Credit_Score`, `Debt_to_Income_Ratio`, and `Credit_Utilization` show a slight positive correlation with `Delinquent_Account`. This is counter-intuitive for `Credit_Score` and `Income`, suggesting that higher values in these might sometimes correlate with delinquency. However, the correlations are very weak.
    - `Loan_Balance`, `Missed_Payments`, and `Account_Tenure` show very weak negative correlations. `Missed_Payments` having a negative correlation is surprising and might indicate a data anomaly or a complex relationship.
- **Categorical Feature Impact**:
    - `Employment_Status`: The 'Unemployed' category appears to have a higher proportion of delinquent accounts compared to other statuses, as might be expected.
    - `Credit_Card_Type`: 'Platinum' and 'Gold' cardholders seem to have a relatively lower proportion of delinquent accounts compared to 'Standard' or 'Student' cardholders, but this needs more in-depth statistical testing.
    - `Location`: There are variations in delinquency rates across different locations, which could be an important regional risk indicator.

These findings suggest that a combination of factors contributes to delinquency, with no single numeric variable having a strong linear relationship. Categorical variables appear to play a more distinguishable role, especially employment status and location.

## 4. Data Quality Issues

### Inconsistent Categorical Values: Employment_Status

```python
# Standardize 'Employment_Status' column
df['Employment_Status'] = df['Employment_Status'].replace({'EMP': 'Employed', 'employed': 'Employed', 'retired': 'Retired'})
display(df['Employment_Status'].value_counts())
```

### Handling Missing Values

```python
# Impute 'Income' and 'Loan_Balance' with their respective medians
df['Income'].fillna(df['Income'].median(), inplace=True)
df['Loan_Balance'].fillna(df['Loan_Balance'].median(), inplace=True)

# Impute 'Credit_Score' with its median
df['Credit_Score'].fillna(df['Credit_Score'].median(), inplace=True)

# Verify no more missing values
print('Missing values after imputation:')
display(df.isnull().sum())
```

### Data Quality Issues Summary

- **Inconsistent Categories**: The `Employment_Status` column had inconsistent entries ('EMP', 'employed', 'retired') which have been standardized to 'Employed' and 'Retired' for better consistency and analysis.
- **Missing Values Imputation**: Missing values in `Income`, `Loan_Balance`, and `Credit_Score` have been imputed with the median of their respective columns. Median imputation was chosen to mitigate the impact of potential outliers, preserving the distribution shape better than mean imputation. After imputation, there are no remaining missing values in the dataset.
